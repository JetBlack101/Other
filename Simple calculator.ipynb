```python
import numpy as np
import tkinter as tk
```

```python
root = tk.Tk()

root.title("Calculator")
root.iconbitmap("calc.ico")
#root.geometry("400x500")

e = tk.Entry(root, width=40, borderwidth=3)

# defining actions
def button_click(number):
    current = e.get()
    e.delete(0,"end")
    e.insert(0, str(current) + str(number))

def button_clr():
    e.delete(0,last="end")

def button_add():
    global symbol
    symbol = "+"
    fno = e.get()
    global f_num
    f_num = fno
    e.delete(0,"end")

def button_sub():
    global symbol
    symbol = "-"
    fno = e.get()
    global f_num
    f_num = fno
    e.delete(0,"end")

def button_mltpl():
    global symbol
    symbol = "*"
    fno = e.get()
    global f_num
    f_num = fno
    e.delete(0,"end")

def button_div():
    global symbol
    symbol = "/"
    fno = e.get()
    global f_num
    f_num = fno
    e.delete(0,"end")    

def button_pwr():
    global symbol
    symbol = "x^y"
    fno = e.get()
    global f_num
    f_num = fno
    e.delete(0,"end")    

def button_sq():
    no = e.get()
    e.delete(0,"end")
    e.insert(0, np.sqrt(float(no)))
    
def button_comma(sep):
    current = e.get()
    e.delete(0,"end")
    e.insert(0, str(current) + str(sep))
    
def button_plsmns(mns):
    current = e.get()
    e.delete(0,"end")
    e.insert(0, str(mns) + str(current))    
    
def button_eq():
    sno =e.get()
    e.delete(0,"end")
    
    if symbol == "+":
        e.insert(0, float(f_num) + float(sno))
    if symbol == "-":
        e.insert(0, float(f_num) - float(sno))
    if  symbol == "*":
        e.insert(0, float(f_num) * float(sno))        
    if symbol == "/":
        e.insert(0, float(f_num) / float(sno))
    if symbol == "x^y":
        e.insert(0, float(f_num) ** float(sno))
        
# defining buttons
button_clear = tk.Button(root,text="Clear", height=2, width=33, command=button_clr, foreground="white", background="#FF791B")

button_1 = tk.Button(root,text="1", height=2, width=6, command=lambda: button_click(1))
button_2 = tk.Button(root,text="2", height=2, width=6, command=lambda: button_click(2))
button_3 = tk.Button(root,text="3", height=2, width=6, command=lambda: button_click(3))
button_4 = tk.Button(root,text="4", height=2, width=6, command=lambda: button_click(4))
button_5 = tk.Button(root,text="5", height=2, width=6, command=lambda: button_click(5))
button_6 = tk.Button(root,text="6", height=2, width=6, command=lambda: button_click(6))
button_7 = tk.Button(root,text="7", height=2, width=6, command=lambda: button_click(7))
button_8 = tk.Button(root,text="8", height=2, width=6, command=lambda: button_click(8))
button_9 = tk.Button(root,text="9", height=2, width=6, command=lambda: button_click(9))
button_0 = tk.Button(root,text="0", height=2, width=6, command=lambda: button_click(0))

button_sep = tk.Button(root,text=",", height=2, width=6, command=lambda: button_comma("."))
button_plusminus = tk.Button(root,text="+/-", height=2, width=6, command=lambda: button_plsmns("-"))

button_divide = tk.Button(root,text="/", height=2, width=6,command=button_div)
button_multipl =  tk.Button(root,text="*", height=2, width=6,command=button_mltpl)
button_minus = tk.Button(root,text="-", height=2, width=6,command=button_sub)
button_plus = tk.Button(root,text="+", height=2, width=6, command=button_add)
button_power2 = tk.Button(root,text="x^y", height=2, width=6, command=button_pwr)
button_sqrt = tk.Button(root,text="sqrt", height=2, width=6, command=button_sq)

button_equal = tk.Button(root,text="=", height=2, width=15, command=button_eq, foreground="white", background="#00806A")

# constructing grid

# row 0
e.grid(row=0, column=0, columnspan=4, padx=2,pady=2)

# row 1
button_clear.grid(row=1,column=0,columnspan=4,  padx=2,pady=2)

# row 2
button_1.grid(row=2,column=0, padx=2,pady=2)
button_2.grid(row=2,column=1, padx=2,pady=2)
button_3.grid(row=2,column=2, padx=2,pady=2)
button_plus.grid(row=2,column=3, padx=2,pady=2)

# row 3
button_4.grid(row=3,column=0, padx=2,pady=2)
button_5.grid(row=3,column=1, padx=2,pady=2)
button_6.grid(row=3,column=2, padx=2,pady=2)
button_minus.grid(row=3,column=3, padx=2,pady=2)

# row 4
button_7.grid(row=4,column=0, padx=2,pady=2)
button_8.grid(row=4,column=1, padx=2,pady=2)
button_9.grid(row=4,column=2, padx=2,pady=2)
button_multipl.grid(row=4,column=3, padx=2,pady=2)

# row 5
button_0.grid(row=5,column=0, padx=2,pady=2)
button_sep.grid(row=5,column=1, padx=2,pady=2)
button_plusminus.grid(row=5,column=2, padx=2,pady=2)
button_divide.grid(row=5,column=3, padx=2,pady=2)

#row 6
button_power2.grid(row=6,column=0, padx=2,pady=2)
button_sqrt.grid(row=6,column=1, padx=2,pady=2)
button_equal.grid(row=6,column=2, columnspan=2, padx=2,pady=2)

root.mainloop()
```
